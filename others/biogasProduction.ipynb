```python
import pandas as pd
from datetime import datetime

# Sample DataFrame with a 'Timestamp' column
data = {
    'Timestamp': pd.date_range(start='2021-01-01', periods=100, freq='H'),
    'Value': range(100)
}
df = pd.DataFrame(data)

# Set 'Timestamp' as the index
df.set_index('Timestamp', inplace=True)

# Resample the DataFrame to hourly intervals and calculate the median
resampled_df = df.resample('H')
median_df = resampled_df.median()

print(median_df)
```

```python
import pandas as pd
import glob
```

```python
csv_files = glob.glob('2022 data/*.csv')
```

```python
combined_df = pd.DataFrame()
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
combined_df
```

```python
combined_df['Data'] = pd.to_datetime(combined_df['Data'])
```

```python
averages = combined_df.groupby(combined_df['Data'].dt.date).mean(numeric_only=True)
```

```python
averages
```

```python
monthly_averages = combined_df.groupby(combined_df['Data'].dt.to_period('M')).mean(numeric_only=True)
```

```python
monthly_averages
```

```python
averages = combined_df.groupby(combined_df['Data'].dt.to_period('Y')).mean(numeric_only=True)
```

```python
averages
```
